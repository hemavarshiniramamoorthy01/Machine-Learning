```python
import pandas as pd
```

```python
dataset = pd.read_csv("50_Startups.csv")
```

```python
dataset.head()
```

```python
datasets=pd.get_dummies(dataset,drop_first=True)
```

```python
datasets.tail()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
num_cols = ['R&D Spend', 'Administration', 'Marketing Spend', 'Profit']
```

```python
scaler = StandardScaler()
```

```python
datasets[num_cols] = scaler.fit_transform(datasets[num_cols])
```

```python
datasets.head()
```

```python
independant = datasets[['R&D Spend','Administration', 'Marketing Spend', 'State_Florida','State_New York']]
```

```python
dependant = datasets[['Profit']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(independant, dependant, test_size = 0.3, random_state = 0)
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel="linear",C=0.1)
regressor=regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
import pickle
```
