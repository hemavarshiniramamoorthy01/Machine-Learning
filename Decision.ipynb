```python
import pandas as pd
```

```python
dataset = pd.read_csv('50_Startups.csv')
```

```python
dataset.head()
```

```python
datasets=pd.get_dummies(dataset,drop_first=True)
```

```python
independant = datasets[['R&D Spend','Administration','Marketing Spend','State_Florida', 'State_New York']]
dependant = datasets[['Profit']]
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independant, dependant, test_size = 1/3, random_state = 0)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion='squared_error', splitter='random')
regressor=regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```
